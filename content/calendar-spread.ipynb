Uma das coisas mais interessantes do mundo de opções financeiras são as estratégias ou [*spreads*](http://en.wikipedia.org/wiki/Options_spread).
De uma forma geral, estratégias são **carteiras** de opções onde um ou mais parâmetros são ajustados para explorar alguma situação específica do mercado, como a alta do ativo objeto, por exemplo.
As estratégias mais conhecidas são:

- *Call Spread* ou *Put Spread*
- *Butterfly Spread*
- *Box Spread*
- *Calendar Spread*

Das estratégias acima o *Calendar Spread* (ou *Spread* Calendário) é a única possui opções com diferentes prazos de vencimento.
Para montar esta estratégia é necessário executar as seguintes operações:

- Venda de uma opção de compra $C_1$ com preço de exercício $K$ e vencimento $T_1$
- Compra de uma opção de compra $C_2$ com preço de exercício $K$ e vencimento $T_2$

onde $T_1 < T_2$.
O custo dessa operação é $-C_1+C_2$.

No vencimento da primeira opção (quando $t = T_1$) temos os seguintes cenários:

- se $S_1 < K$ então $C_1 = 0 \longrightarrow C_2$
- se $S_1 > K$ então $C_1 > 0 \longrightarrow -S_{T_1} + K + C_2$

No primeiro cenário com certeza temos uma carteira positiva, dado que $C_2 > 0$, no entanto, no segundo cenário, onde $C_1$ é exercída, não podemos afirmar se a carteira resultante $-S_{T_1} + K + C_2$ é positiva.

Pelo apreçamento neutro ao risco temos:

$$
C_2 = E^Q \left[\max(S_2 - K, 0)\,|\, F_{T_1} \right]
$$

e aplicando o lema de Jensen a equação acima temos:

$$
\begin{split}
E^Q \left[ \max(S_2 - K, 0)\,|\, F_{T_1} \right] &\ge \max(E^Q[S_2\,|\, F_{T_1}] - K, 0) \\
& = \max(S_1 - K, 0)
\end{split}
$$

Para o cenário em que $S_1 > K$ temos que:

$$
C_2 = E^Q \left[ \max(S_2 - K, 0) \,|\, F_{T_1} \right] \ge S_1 - K
$$

Portanto

$$
-S_1 + K + C_2>0
$$

```python
from numpy import sqrt
from scipy import exp, log
from scipy.stats import norm

def d1d2(S, K, T, r, v, q):
	d1 = ( log(S/K) + (r - q + 0.5*v**2)*T ) / (v*sqrt(T))
	d2 = d1 - v*sqrt(T)
	return d1, d2


def put(S, K, T, r, v, q):
	d1, d2 = d1d2(S, K, T, r, v, q)
	N1 = norm.cdf(d1)
	N2 = norm.cdf(d2)
	
	return -S*exp(-q*T)*(1-N1) + K*exp(-r*T)*(1-N2)


def call(S, K, T, r, v, q):
	d1, d2 = d1d2(S, K, T, r, v, q)
	N1 = norm.cdf(d1)
	N2 = norm.cdf(d2)
	
	return S*exp(-q*T)*N1 - K*exp(-r*T)*N2

def delta(S, K, T, r, v, q):
    d1, d2 = d1d2(S, K, T, r, v, q)
    return exp(-q*T)*norm.cdf(d1)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
S = np.linspace(70, 150, 1000)
```

```python
plt.plot(S, call(S, 110, 3/12, 0.1, 0.2, 0)-call(S, 110, 1/12, 0.1, 0.2, 0))
plt.axvline(110, linestyle="dashed", color="grey")
plt.ylim(0, 5)
plt.title("Calendar Spread");
```

```python
plt.plot(S, call(S, 110, 2/12, 0.1, 0.2, 0) - np.max(np.vstack((S-110, np.zeros(S.shape))), 0))
plt.axvline(110, linestyle="dashed", color="grey")
plt.ylim(0, 5)
plt.title("Calendar Spread");
```
